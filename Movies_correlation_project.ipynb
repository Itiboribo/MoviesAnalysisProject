What are the 10 most and least profitable movies?
over the period timeframe without considering the economic and social factors like inflation and impact of social media awareness. 

```python
#adding a new column called profit 
# profit depends on the revenue and the budget
df2['profit']= df2['gross'] - df2['budget']
df2.head()
```

```python
pd.DataFrame(df2['profit'].sort_values(ascending = False))
```

```python
#making a plot to show top 10 most profitable movies.
#sorting the 'Profit' column in decending order and store it in a new dataframe,
profit_df = pd.DataFrame(df2['profit'].sort_values(ascending = False))

#creating new column 'original_title' in the new dataframe from the original dataframe
profit_df['name'] = df2['name']
x = list(map(str,(profit_df['name'])))
x = list(x[:10])
y = list(profit_df['profit'][:10])

#make a plot using pointplot for top 10 profitable movies.
ax = sns.pointplot(x=y,y=x)

#setup the figure size
sns.set(rc={'figure.figsize':(8,5)})
#setup the title and labels of the plot.
ax.set_title("Top 10 Profitable Movies",fontsize = 15)
ax.set_xlabel("Profit",fontsize = 13);
```

```python
profit_df
```

```python
#10 least profitable movies.

profit_df_l = pd.DataFrame(df2['profit'].sort_values(ascending = True))

#creating new column 'original_title' in the new dataframe from the original dataframe
profit_df_l['name'] = df2['name']
x = list(map(str,(profit_df_l['name'])))
x = list(x[:10])
y = list(profit_df_l['profit'][:10])

#making a plot using pointplot for top 10 profitable movies.
ax = sns.pointplot(x=y,y=x)

#setup the figure size
sns.set(rc={'figure.figsize':(8,5)})

#setup the title and labels of the plot.
ax.set_title("Least 10 Profitable Movies",fontsize = 15)
ax.set_xlabel("Profit",fontsize = 13);
```

Analysis of Relation between Budgets and Gross from Movies
What are the correlations between gross, budget and profit?

```python
df2.columns
```

```python
# correlation between budget and gross
# scatter plot with budget vs gross 

plt.scatter(x=df2['budget'], y=df2['gross'])
plt.title('Budget vs Gross')
plt.xlabel(' Film Budget')
plt.ylabel('Gross Earning')
plt.show
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df2, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
plt.scatter(x=df2['profit'], y=df2['profit'])
plt.title('gross vs profit')
plt.xlabel(' Film gross')
plt.ylabel('profit')
plt.show
```

```python
#types of correlations: pearson, kendall,spearman
df2.corr(method='pearson')
```

```python
correlation_matrix = df2.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.show()
```

from a research papaer, 
weak positive correlation would be in the range of 0.3 to 0.5,
moderate positive correlation from 0.5 to 0.70, 
strong positive correlation from 0.7 to 0.99 
perfect correlation is 1

The correlation between Budget and gross is found to be 0.74 which is a strong positive correlation between the variables. This means that with increase in budget there is an increase in the Gross.
Also, with a corralation value of 0.98, the relationship between profit and gross is linear.
